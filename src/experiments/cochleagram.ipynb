```python
import pycochleagram.cochleagram as cgram
from pycochleagram.utils import wav_to_array
import os
import numpy as np
import re
import cv2
from scipy.io import wavfile
import matplotlib.pyplot as plt
from scipy import signal
```

```python
def generate_cochleagram(wav_file):
    sig, sr = wav_to_array(wav_file)
    sig = sig[:50000]
    print(sig.shape, sr)
    coch = cgram.human_cochleagram(sig, sr, n=150, sample_factor=1, downsample=None, nonlinearity='db')
    img = np.flipud(coch)
    return img
```

```python
coch = generate_cochleagram('../data/vis-data-256/2015-02-16-17-08-51_denoised.wav')
plt.imshow(coch, aspect='auto', cmap='magma', origin='lower', interpolation='nearest')
```
